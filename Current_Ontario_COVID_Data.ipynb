## Code Generated from OpenAI

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load csv data 
covid_data = pd.read_csv("https://health-infobase.canada.ca/src/data/covidLive/covid19-download.csv")

# Filter for Ontario
ontario_data = covid_data[covid_data.prname == "Ontario"]

# Extract avgdeaths_last7
avgdeaths_last7 = ontario_data.avgdeaths_last7

# Plot avgdeaths_last7
plt.plot(avgdeaths_last7)
plt.title('Average Deaths in Last 7 Days in Ontario')
plt.xlabel('Date')
plt.ylabel('Average Deaths')
plt.show()
```
