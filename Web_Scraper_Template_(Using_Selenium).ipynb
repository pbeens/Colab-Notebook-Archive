```python
# setup takes a long time

!pip install selenium -q
!apt-get update # to update ubuntu to correctly run apt install
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
from selenium.webdriver.common.by import By
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
# global variable(s)
url = 'https://www.energy.gov/eere/solar/how-does-solar-work'
```

```python
wd.get(url)

time.sleep(0.5)

links = wd.find_elements_by_xpath("//a[@href]") # finds all links

text = wd.find_element_by_xpath("/html/body").text
```

```python
# this section gets links

links_dict = {}

for link in links:
    title = link.text
    url = link.get_attribute('href')
    links_dict[url] = title

for item in sorted(links_dict.items()):
    print(item)
```

```python
# this section counts and prints the words 

all_words = {}

for word in text.split():
    if word not in all_words:
        all_words[word] = 1
    else:
        all_words[word] += 1

sorted_dict = dict(sorted(all_words.items(), reverse=True, key=lambda item: item[1]))
for word, count in sorted_dict.items():
    print(f'{word}: {count}')

```
